```python
import json
from tqdm import tqdm
import pandas as pd
```

```python
with open('captions_val2017.json') as f:
    jfile = json.load(f)
```

```python
jfile['images']
```

```python
image_ids = []
file_paths = []
images_urls = []
for image in tqdm(jfile['images']):
    image_ids.append(image['id'])
    file_paths.append(image['file_name'])
    images_urls.append(image['coco_url'])
assert len(image_ids) == len(file_paths)
```

```python
df2 = pd.DataFrame({'image_id': image_ids, 'file_path': file_paths, 'image_url': images_urls})
df2.head()
```

```python
dataset = pd.read_excel('arabicf.xls', header=None).fillna(0)
dataset.head(7)
```

```python
all_caps = []
ids = []
for row in dataset.itertuples():
    if row[1] == 0:
        continue
    all_caps.append(row[1])
    if row[2] != 0:
        c_id = row[2].split('=')[1]
    ids.append(c_id)
```

```python
sm_dataset = pd.DataFrame({'caption':all_caps, 'id': ids})
sm_dataset.id = pd.to_numeric(sm_dataset.id)
sm_dataset.head(11)
```

```python
df_image_1 = df2[df2.image_id.isin(sm_dataset.id.unique())]
```

```python
df_image_2 = df2[df2.image_id.isin(sm_dataset.id.unique())]
```
