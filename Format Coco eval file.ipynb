```python
import json
from tqdm import tqdm
import pandas as pd
```

```python
with open('data/captions_val2014.json') as f:
    jfile = json.load(f)
```

```python
image_ids = []
captions = []
for ann in tqdm(jfile['annotations']):
    image_ids.append(ann['image_id'])
    captions.append(ann['caption'])
```

```python
df1 = pd.DataFrame({'image_id': image_ids, 'captions': captions})
df1.head()
```

```python
jfile['images']
```

```python
image_ids = []
file_paths = []
for image in tqdm(jfile['images']):
    image_ids.append(image['id'])
    file_paths.append(image['file_name'])
assert len(image_ids) == len(file_paths)
```

```python
df2 = pd.DataFrame({'image_id': image_ids, 'file_path': file_paths})
df2.head()
```

```python
full_df = df1.merge(df2)
full_df.head(11)
```
